```python
import pandas as pd     #importing the libraries
import matplotlib.pyplot as plt     #importing the libraries
import numpy as np          #importing the libraries
import seaborn as sns                   #importing the libraries
from sklearn.model_selection import train_test_split    #importing the libraries
from sklearn.linear_model import LinearRegression      #importing the libraries
from sklearn.metrics import r2_score 
from sklearn.metrics import mean_absolute_error 
```

```python
data = pd.read_csv("data1.csv")   # read the data from the csv file
```

```python
data                  #display the data
```

```python
data.isna().sum() #check for missing values
```

```python
data.describe() #display the summary statistics
```

```python
a = data.drop("Salary", axis = 1) #drop the salary column
```

```python
b = data.Salary
```

```python
b
```

```python
plt.scatter(a,b)                        #scatter plot
```

```python
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size = 0.2, random_state = 0) #split the data into training and testing data
```

```python
b_train                                                         #display the training data
```

```python
lre = LinearRegression() #create a linear regression model
lre
```

```python
lre
```

```python
lre.fit(a_train,b_train) #fit the model
```

```python
bpred = lre.predict(a_test)
```

```python
bpred
```

```python
b_test
```

```python
r2_score(bpred,b_test)
```

```python
mean_absolute_error(bpred,b_test)
```

```python
sns.regplot(x=a_test, y=b_test)  # Régression plot
```
